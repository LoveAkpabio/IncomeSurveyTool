```python
# Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set a theme for Seaborn to make visualizations look consistent
sns.set_theme(style="whitegrid")

# Step 2: Load the user data from the CSV file
data_path = '../data/user_data.csv'  # Adjust the path if necessary

try:
    df = pd.read_csv(data_path)
    print("Data loaded successfully!\n")
    print(df.head())  # Show the first few rows to verify data
except FileNotFoundError:
    print(f"Error: The file at {data_path} was not found.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")

# Step 3: Visualize Income by Age Group using a bar plot
plt.figure(figsize=(8, 5))  # Set figure size
sns.barplot(data=df, x='Age', y='Income', hue='Age', errorbar=None, palette='muted', legend=False)  
plt.title('Income by Age Group')  # Add a title
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.legend([], [], frameon=False)  # Hide legend to comply with API change
plt.show()  # Display the plot

# Step 4: Melt the DataFrame to plot spending categories per gender
melted_df = df.melt(id_vars='Gender', 
                    value_vars=['Utilities', 'Entertainment', 'School Fees', 'Shopping'], 
                    var_name='Category', value_name='Amount')

plt.figure(figsize=(10, 6))  # Set figure size
sns.barplot(data=melted_df, x='Category', y='Amount', hue='Gender', errorbar=None)  # Create bar plot
plt.title('Gender Distribution Across Spending Categories')  # Add a title
plt.show()  # Display the plot

# Step 5: Calculate Total Expenses by summing selected categories
df['Total_Expenses'] = df[['Utilities', 'Entertainment', 'School Fees', 'Shopping']].sum(axis=1)

# Visualize Total Expenses by Gender using a bar plot
plt.figure(figsize=(8, 5))  # Set figure size
sns.barplot(data=df, x='Gender', y='Total_Expenses', hue='Gender', errorbar=None, palette='coolwarm', legend=False)  
plt.title('Total Expenses by Gender')  # Add a title
plt.legend([], [], frameon=False)  # Hide legend to comply with API change
plt.show()  # Display the plot

# Step 6: Save the plots as PNG files (for use in presentations or reports)
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='Age', y='Income', hue='Age', errorbar=None, palette='muted', legend=False)
plt.title('Income by Age Group')
plt.xticks(rotation=45)
plt.legend([], [], frameon=False)  # Hide legend
plt.savefig('../data/income_by_age.png')  # Save plot as PNG

plt.figure(figsize=(10, 6))
sns.barplot(data=melted_df, x='Category', y='Amount', hue='Gender', errorbar=None)
plt.title('Gender Distribution Across Spending Categories')
plt.savefig('../data/gender_spending.png')  # Save plot as PNG

# Final message
print("Visualizations generated and saved successfully!")
```
